```python
import pandas as pd
```

```python
clientes=pd.read_csv('clientes.csv',index_col=0)
clientes
```

```python
cuentas=pd.read_csv('cuentas.csv',index_col=0)
cuentas
```

```python
movimientos=pd.read_csv('movimientos.csv',index_col=0)
movimientos
```

```python
import sqlite3
conn = sqlite3.connect('old.db')
c = conn.cursor()
```

```python
clientes.to_sql('clientes', conn, if_exists='replace', index = False)
cuentas.to_sql('cuentas', conn, if_exists='replace', index = False)
movimientos.to_sql('movimientos', conn, if_exists='replace', index = False)
```

```python
prestamos = pd.DataFrame(columns=['cliente_id','monto','fecha_inicio','tipo'])
```

```python
prestamos.to_sql('prestamos', conn, if_exists='replace', index = False)
```
